# Instalando Git e Github Desktop

Link para Baixar o Git:
https://git-scm.com/

Já na pagina inicial você encontrara as opções de download

É durante a instação que você indica qual editor utiliza, atente-se a isso!

O restante é só next, next, next...

link pra baixar github desktop:
https://desktop.github.com/

Já na tela inicial você encontra a opção de download

A instalação é feita de forma totalmente autmatica, basta abrir o executavel
